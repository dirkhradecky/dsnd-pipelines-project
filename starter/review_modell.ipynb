## The Data

The dataset has been anonymized and cleaned of missing values.

There are 8 features for to use to predict whether a customer recommends or does
not recommend a product.
The `Recommended IND` column gives whether a customer recommends the product
where `1` is recommended and a `0` is not recommended.
This is your model's target/

The features can be summarized as the following:

- **Clothing ID**: Integer Categorical variable that refers to the specific piece being reviewed.
- **Age**: Positive Integer variable of the reviewers age.
- **Title**: String variable for the title of the review.
- **Review Text**: String variable for the review body.
- **Positive Feedback Count**: Positive Integer documenting the number of other customers who found this review positive.
- **Division Name**: Categorical name of the product high level division.
- **Department Name**: Categorical name of the product department name.
- **Class Name**: Categorical name of the product class name.

The target:
- **Recommended IND**: Binary variable stating where the customer recommends the product where 1 is recommended, 0 is not recommended.

#### Character Count

```python
from sklearn.base import BaseEstimator, TransformerMixin
class CountCharacter(BaseEstimator, TransformerMixin):
    def __init__(self, character: str):
        self.character = character

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return [[text.count(self.character)] for text in X]
```

#### Review Length (Word Count)

```python
class WordCount(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        word_counts = [len(str(text).split()) for text in X]
        return np.array(word_counts).reshape(-1, 1)
```

#### Ratio Uppercase

```python
class UppercaseRatio(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        ratios = []
        for text in X:
            text = str(text)
            total_chars = len(text)
            if total_chars == 0:
                ratios.append(0)
            else:
                uppercase_chars = sum(1 for c in text if c.isupper())
                ratios.append(uppercase_chars / total_chars)
        return np.array(ratios).reshape(-1, 1)
```

```python
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import FunctionTransformer
import numpy as np
from scipy.sparse import csr_matrix

class JoinCols(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return (X.iloc[:, 0].fillna('') + ' ' + X.iloc[:, 1].fillna('')).values

class ToCsr(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return csr_matrix(X)

initial_text_preprocess = Pipeline([
    ('column_joiner', JoinCols()),
])

feature_engineering = FeatureUnion([
    ('count_spaces', CountCharacter(character=' ')),
    ('count_exclamations', CountCharacter(character='!')),
    ('count_question_marks', CountCharacter(character='?')),
    ('review_length', WordCount()),
    ('uppercase_ratio', UppercaseRatio()),
])

character_counts_pipeline = Pipeline([
    ('initial_text_preprocess', initial_text_preprocess),
    ('feature_engineering', feature_engineering),
    ('to_csr', ToCsr()),
])
character_counts_pipeline
```

#### spaCy and TF-IDF

```python
class SpacyLemmatizer(BaseEstimator, TransformerMixin):
    def __init__(self, nlp):
        self.nlp = nlp

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        lemmatized = [
            ' '.join(
                token.lemma_ for token in doc
                if not token.is_stop
            )
            for doc in self.nlp.pipe(X)
        ]
        return lemmatized
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_pipeline = Pipeline([
    ('column_joiner', JoinCols()),
    ('lemmatizer', SpacyLemmatizer(nlp=nlp)),
    ('tfidf_vectorizer', TfidfVectorizer(stop_words='english')),
])
tfidf_pipeline 
```

#### Sentiment Analysis

```python
import torch
```

```python
from transformers import pipeline as hf_pipeline
class SentimentAnalyzer(BaseEstimator, TransformerMixin):
    def __init__(self, model_id="distilbert-base-uncased-finetuned-sst-2-english",
                 device=0, batch_size=32, max_length=128):
        self.model_id = model_id
        self.device = device
        self.batch_size = batch_size
        self.max_length = max_length
        self._pipe = None

    def _ensure_pipe(self):
        if self._pipe is None:
            self._pipe = hf_pipeline(
                "sentiment-analysis",
                model=self.model_id,
                device=self.device
            )

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        self._ensure_pipe()
        texts = list(X)
        
        out = self._pipe(
            texts,
            batch_size=self.batch_size,
            truncation=True,
            max_length=self.max_length
        )
        scores = [o["score"] if o["label"] == "POSITIVE" else -o["score"] for o in out]
        return np.asarray(scores).reshape(-1, 1)
```

```python
sentiment_pipeline = Pipeline([
    ('initial_text_preprocess', initial_text_preprocess),
    ('sentiment_analyzer', SentimentAnalyzer())
])

sentiment_pipeline
```

## Training Pipeline

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
from joblib import Memory

memory = Memory("./cache", verbose=0)

clf = LogisticRegression(max_iter=2000, solver="saga", class_weight="balanced")

model_pipeline_lr = Pipeline([
    ('feature_engineering', feature_engineering),
    ('clf', clf)
], memory=memory
)

model_pipeline_lr
```

```python
pipe = Pipeline([
    ("feature_engineering", feature_engineering),
    ("clf", LogisticRegression())   # Dummy; wird von param_grid überschrieben
], memory=memory
)

param_grids = [
    {   # Logistic Regression
        "clf": [LogisticRegression(max_iter=2000, solver="saga")],
        "clf__C": [0.25, 0.5, 1, 2],
        "clf__class_weight": ["balanced", None],
        "clf__penalty": ["l2"],
    },
    {   # Linear SVC
        "clf": [LinearSVC()],
        "clf__C": [0.25, 0.5, 1, 2],
        "clf__loss": ["hinge", "squared_hinge"],
        "clf__class_weight": ["balanced", None],
    },
]

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

gs = GridSearchCV(
    pipe,
    param_grid=param_grids,
    cv=cv,
    scoring={
        "accuracy": "accuracy",
        "f1_macro": "f1_macro",
        "bal_acc": "balanced_accuracy",
        "roc_auc": "roc_auc"
    },
    refit="f1_macro",
    n_jobs=1,
    verbose=2,
    error_score="raise"
)

gs.fit(X_train, y_train)
print("Best model:", gs.best_estimator_["clf"])
print("Best params:", gs.best_params_)
print("Best f1_macro:", gs.best_score_)
```
